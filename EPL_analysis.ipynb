**DECODING GOAL PATTERNS: A DATA DRIVEN ANALYSIS OF THE EPL 2023/24 SEASON**


```python
#i am importing all the generic models for the data loading and the data exploration
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the dataset and the testing of it

df=pd.read_csv('/content/matches.csv')
df.describe()
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.isnull()
```

```python
# convert to from date to datetime
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
#dropping columns that are not much needed
#dropping the unnamed, Comp, day, match report,notes,poss,attendence.

df.drop(columns=["Match Report", "Notes","Comp",'Attendance','Captain',"Formation","Referee","Unnamed: 0"], inplace=True)
```

```python
df.head()
```

```python
df.drop(columns=["Season"],inplace=True)
```

```python
df.isna()
```

```python
#now for the data visualisation part of it
#the plot of goal distibution using a scatterplot

team_results = df.groupby("Team")["Result"].value_counts().unstack().fillna(1)
team_results.plot(kind="bar", stacked=True, figsize=(12, 6), colormap="viridis")

plt.xlabel("Team")
plt.ylabel("Number of Matches")
plt.title("Wins, Draws, and Losses per Team in EPL 2023/24")
plt.xticks(rotation=90)
plt.legend(["Losses", "Draws", "Wins"])
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Inference:
# It seems that Manchester City has won more matches than any other team as well as the lowestt number of draws and losses
# On the other hand, Sheffield United has the less number of wins but the most number of draws
# The team with the highest number of losses is Brighton and Hove Albion
```

```python
#Number of Home vs Away wins


# Count home and away wins
home_wins = df[(df['Venue'] == 'Home') & (df['Result'] == 'W')].shape[0]
away_wins = df[(df['Venue'] == 'Away') & (df['Result'] == 'W')].shape[0]

# Bar chart
plt.figure(figsize=(6, 4))
sns.barplot(x=['Home Wins', 'Away Wins'], y=[home_wins, away_wins], palette=['pink', 'cyan'])
plt.ylabel("Number of Wins")
plt.title("Home Wins vs Away Wins in EPL 2023/24")
plt.show()

#Inference
# It seems that many teams are winning in their home turf rather than in away games
```

```python
#KDE graph for Expected Goals(xG) versus Actual Goals(GF)

plt.figure(figsize=(10,6))
sns.kdeplot(df['xG'], shade=True, label='Expected Goals (xG)', color='red')
sns.kdeplot(df['GF'], shade=True, label='Actual Goals (GF)', color='green')
plt.xlabel('Goals')
plt.ylabel('Density')
plt.title('Density Distribution of xG vs GA')
plt.grid(True)
plt.legend()
plt.show()

#Inference
#This KDE graph shows that the players are overperforming during each match ie goals are being scored more than the expected goals for most teams.
```

```python
#KDE graph for Expected Goals conceded (xGA) versus Actual Goals Conceded(GA)

plt.figure(figsize=(10,6))
sns.kdeplot(df['xGA'], shade=True, label='Expected Goals conceded (xGA)', color='red')
sns.kdeplot(df['GA'], shade=True, label='Actual Goals conceded (GA)', color='BLUE')
plt.xlabel('Goals')
plt.ylabel('Density')
plt.title('Density Distribution of xG vs GA')
plt.grid(True)
plt.legend()
plt.show()

#Inference
#This KDE graph shows that there were actually lesser goals conceded than the expeceted goals to be conceded
```

```python
#Goals scored for their respective teams
plt.figure(figsize=(12,6))
for team in df['Team'].unique():
    team_df = df[df['Team'] == team]
    plt.plot(team_df['Date'], team_df['GF'].cumsum(), label=team)

plt.xlabel('Date')
plt.ylabel('Cumulative Goals')
plt.title('Goals Scored Over the Season')
plt.legend()
plt.grid(True)
plt.show()

#inference
#this line chart shows that Manchester City has scored the most goals as well as they were the winners of the EPL 2023/24 season
#On the other hand, Sheffield United has scared the least amount of goals
```

```python
#Goals conceded  by the team using a Line chart
plt.figure(figsize=(12,6))
for team in df['Team'].unique():
    team_df = df[df['Team'] == team]
    plt.plot(team_df['Date'], team_df['GA'].cumsum(), label=team)

plt.xlabel('Date')
plt.ylabel('Cumulative Goals')
plt.title('Goals conceded Over the Season')
plt.legend()
plt.grid(True)
plt.show()

#inference
#this line chart shows us the team with the most goals conceded is Sheffield United
#However the less number of goals conceded is Arsenal
```

```python
# Calculate successful and missed penalty kicks
penalty_data = {
    'Penalties scored': df['PK'].sum(),
    'Missed penalties': (df['PKatt'] - df['PK']).sum()
}

# Plot the pie chart
plt.figure(figsize=(8,8))
plt.pie(penalty_data.values(), labels=penalty_data.keys(), autopct='%1.1f%%', colors=['green', 'red'])
plt.title('Penalty Kick Success vs. Missed Attempts')
plt.show()

#Inference
#this below pie chart shows us that nearly 90% of the penaltiies are scored
#this also implies that many penalty kicks are a mak=jor way to score goals
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Team', y='Dist', data=df)
plt.xticks(rotation=90)
plt.xlabel('Team')
plt.ylabel('Average Shot Distance (Dist)')
plt.title('Shot Distance Distribution by Team')
plt.show()

#Inference
#Teams like Manchester City and Arsenal take closer-range shots, indicating a more intricate attacking style, while
#Sheffield United and Burnley attempt longer-range shots, reflecting a more direct approach.
#Consistent shot distances in teams like Liverpool suggest structured play, whereas larger spreads in teams like Wolverhampton indicate varied attacking strategies.
```
